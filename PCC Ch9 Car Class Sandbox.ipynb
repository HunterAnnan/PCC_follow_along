```python
#define the broad class "car"

class Car():
    """Broad car class"""
    
    def __init__(self, make, model, year):
        self.make = make
        self.model = model
        self.year = year
        #full name using above info
        self.fullname = str(self.year) + " " + self.make + " " + self.model
        self.odometer = 0
        
#    def getname(self):
#        fullname = str(self.year) + " " + self.make + " " + self.model
#        return fullname.title()
    
    def read_odometer(self):
        print(
            "The odometer reads",
            self.odometer,
            "km."
        )
        
class ICECar(Car):
    """Car child class for ICE cars"""
    
    def __init__(self, make, model, year, fuel_range, fuel_level = 0):
        super().__init__(make, model, year)
        self.fuel_range = fuel_range
        self.fuel_level = fuel_level
        
    def get_fuel_level(self):
        print(
            "Charge: " +
            str(round(self.fuel_level*100, 1)) +
            "%"
        )
        
    def refuel(self):
        fuel_i = self.fuel_level
        print(
            "Refueled " +
            str(round(((1-fuel_i)*100), 1)) +
            "%; now full."
        )
        self.fuel_level = 1
        
    def drive(self, dist, speed=50):
        drag_eff = -(1/200)*speed + 1.25
        usage = ((dist/self.fuel_range)/drag_eff)
        
        if self.fuel_level < usage:
            print("You won't make it: adjust your speed and/or distance.")
        else:
            self.fuel_level -= usage
            self.odometer += dist
            print("You travel " + str(dist) + " km.")
    
        
class ElectricCar(Car):
    """Car child class for electric cars"""
    
    def __init__(self, make, model, year, bat_range, bat_charge = 0, bat_cycles = 0):
        super().__init__(make, model, year)
        self.bat_range = bat_range
        self.bat_charge = bat_charge
        self.bat_cycles = bat_cycles
        
    def get_charge(self):
        print(
            "Charge: " + 
            str(round(self.bat_charge*100, 1)) +
            "%"
        )
        
    def get_cycles(self):
        print(
            "Battery cycles: " + 
            str(round(self.bat_cycles, 1))
        )
        
    def charge(self):
        chg_i = self.bat_charge
        print(
            "Charged " + 
            str(round(((1-chg_i)*100), 1)) +
            "%; now full."
        )
        self.bat_charge = 1
        
    def drive(self, dist, speed=50):
        drag_eff = -(1/200)*speed + 1.25
        usage = ((dist/self.bat_range)/drag_eff)
        
        if self.bat_charge < usage:
            print("You won't make it: adjust your speed and/or distance.")
        else:
            self.bat_cycles += usage
            self.bat_charge -= usage
            self.odometer += dist
            print("You travel " + str(dist) + " km.")

mycar = ICECar('mazda', 'MX5', 2009, 563)
myev = ElectricCar('tesla', 'model s', 2020, 651)

print("Topping off both cars:")
mycar.refuel()
myev.charge()
print("")

print("My ICE car:")
mycar.drive(300, 50)
mycar.get_fuel_level()
mycar.refuel()
mycar.read_odometer()
print("")

print("My EV:")
myev.drive(300, 50)
myev.get_charge()
myev.charge()
myev.read_odometer()

myev.get_cycles()
```

```python
# Accept user inputs to choose what they would like to do, from:
## Defining new cars
## Interacting with those cars

### Start with zero cars

car_counter = 0

cars = {}

### Define some prompts

prompt_top =  """What would you like to do? Type "q" at any time to quit.
a. Add a new car to your garage, or
b. Interact with your cars.\n"""

prompt_a = """Add a new car: enter a type below.
ICE: \tinternal combustion
EV: \telectric motor\n"""

prompt_a_ice_1 = "Internal combustion car: enter the brand below.\n"
prompt_a_ev_1 = "Electric car: enter the brand below.\n"
prompt_a_2 = "Enter the model below.\n"
prompt_a_3 = "Enter the year below.\n"
prompt_a_4 = "What is the range, in kilometers, of this car?\n"

prompt_b = "Interact: enter a car to interact with.\n"

#MASTER LOOP, with
#top level input: choose to add new car, or interact
while True:
    i = input(prompt_top)
    
    #TOP > quit
    if i.lower() == "q":
        break
        
    #TOP > Add new car
    elif i.lower() == "a":
        i = input(prompt_a)
        
        #TOP > Add new car > QUIT
        if i.lower() == "q":
            print("quit test")
            
        #TOP > Add new car > ICE
        elif i.lower() == "ice":
            temp_classtype = ICECar

            temp_classmake = input(prompt_a_ice_1)
            if temp_classmake == "q":
                break
            temp_classmake = temp_classmake.lower()

            temp_classmodel = input(prompt_a_2)
            if temp_classmodel == "q":
                break
            temp_classmodel = temp_classmodel.lower()

            temp_classyear = int(input(prompt_a_3))
            if temp_classyear == "q":
                break
            temp_classfuelrange = int(input(prompt_a_4))
            if temp_classfuelrange == "q":
                break

            if car_counter < 2:
                car_counter += 1
                cars["car_" + str(car_counter)] = ICECar(
                    temp_classmake,
                    temp_classmodel,
                    temp_classyear,
                    temp_classfuelrange
                )
                print(
                    cars["car_" + str(car_counter)].fullname.title() 
                    + " with " 
                    + str(cars["car_" + str(car_counter)].fuel_range) 
                    + " km of range added to your garage.")
                prompt_b += str(car_counter) + ": " + cars["car_" + str(car_counter)].fullname.title() + "\n"
            elif car_counter >= 3:
                print("Error: too many cars...")
            
        #TOP > Add new car > EV
        elif i.lower() == "ev":
            temp_classtype = ElectricCar

            temp_classmake = input(prompt_a_ev_1)
            temp_classmake = temp_classmake.lower()

            temp_classmodel = input(prompt_a_2)
            temp_classmodel = temp_classmodel.lower()

            temp_classyear = int(input(prompt_a_3))
            temp_classfuelrange = int(input(prompt_a_4))

            if car_counter < 2:
                car_counter += 1
                cars["car_" + str(car_counter)] = ElectricCar(
                    temp_classmake, 
                    temp_classmodel, 
                    temp_classyear, 
                    temp_classfuelrange
                )
                print(
                    cars["car_" + str(car_counter)].fullname.title() 
                    + " with " 
                    + str(cars["car_" + str(car_counter)].bat_range) 
                    + " km of range added to your garage.")
                prompt_b += str(car_counter) + ": " + cars["car_" + str(car_counter)].fullname.title() + "\n"
            elif car_counter >= 3:
                print("Error: something went wrong...")

        #TOP > Add new car > Unknown type (error)
        else:
            print("Invalid vehicle type.")
    
    #TOP > Interact (no cars)
    elif i.lower() == "b" and car_counter == 0:
        print("You don't have any cars to interact with.")
    
    #TOP > Interact (with cars)
    elif i.lower() == "b":
        i = input(prompt_b)
        
        #TOP > Interact > QUIT
        if i.lower() == "q":
            break
            
        #TOP > Interact > Check valid
        elif isinstance(int(i), int) and int(i) <= car_counter:
            
            #TOP > Interact > Checked valid > 
            interact = input(i + ": " + cars["car_" + str(i)].fullname.title()) + "\n"
            if isinstance(cars["car_" + str(i)], ElectricCar):
                print("Test/placeholder EV interaction.")
            elif isinstance(cars["car_" + str(i)], ICECar):
                print("Test/placeholder ICE interaction.")
        
        #TOP > Interact > Invalid entry (error)
        else:
            print("Invalid selection.")
```
